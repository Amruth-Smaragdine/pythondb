```python
# Code to print numbers from 1 to 9. 
# This is a manual method where multiple print for all numbers in sequence
print (1)
print (2)
print (3)
print (4)
print (5)
print (6)
print (7)
print (8)
print (9)
```

```python
# Code to automate printing of a number using while loop.
# In this example we are initiating a value with '1' before the while loop
# We have set the higher limit as 10 in the while loop
# We are trying to print the value of count and increment it by one every time.
# The moment condition of "count is less than (<) 10" then the execution stops.
# In the same code if we place "count is less than (or) equal to 10 (<=)" then '10' also gets printed
# Warning: Please comment one of the 'while' statement
count = 1
while count < 10:
# while count <= 10:
    print ("Value of 'count' variable is:", count)
    count = count + 1
```

```python
# The same example can be repeated by using 'for' loop.
# While using for loop we can include 'count' variable in the same statement.
# The statement 'range' takes three values (Initial value, Maximum value and Increment value)
# Normally in all the for statements we only see 'range' with 'Initial value' and 'Maximum value'
# By default the 'Increment Value' is '1'
# NOTE: In all the range statements of for loop the 'Initial value' is included. The 'Maximum value' is excluded.
#       Which is the reason that 10 does not get printed.
for count in range (1, 10):
    print ("Value of 'count' variable is:", count)
```

```python
# This is a nested for loop. We made use of two variables 'a' and 'b'
# When the code is executed for the first time a = 1 and b will be assigned values 1, 2, 3, . . . 9
# When the code is executed for the second time a = 2 and b will once again get values 1, 2, 3, . . . 9
# Similarly the code will execute until the value of a reaches 9
for a in range (1, 11):
    for b in range (1, 11):
        print ("Value of a:", a, "Value of b:", b)
```

```python
# We can extend the above code and can print addition, subtraction, multiplication and division tables as follows:

print ("Addition Table")
for a in range (1, 11):
    for b in range (1, 11):
        print ("{} + {} = {}".format(a, b, a + b))
print ("\n")
print ("Subtraction Table")
for a in range (1, 11):
    for b in range (1, 11):
        print ("{} - {} = {}".format(a, b, a - b))
print ("\n")
print ("Multiplication Table")
for a in range (1, 11):
    for b in range (1, 11):
        print ("{} x {} = {}".format(a, b, a * b))
print ("\n")
print ("Division Table")
for a in range (1, 11):
    for b in range (1, 11):
        print ("{} / {} = {}".format(a, b, a / b))
```
