```python
# First Version
# Code snippet to extract value from Microsoft excel workbook
# Cells with no data are intentionally marked with an 'x'

import xlrd
WorkBook = xlrd.open_workbook("SampleWorkbook.xlsx")
WorkSheet = WorkBook.sheet_by_index(0)

for i in range (0, WorkSheet.nrows):
    for j in range (0, WorkSheet.ncols):
        if WorkSheet.cell_value(i,j) == '':
            print ("x")
        else: 
            print (WorkSheet.cell_value(i, j))
    print ("\n")
```

```python
# First Version - Iteration - 02

import xlrd
WorkBook = xlrd.open_workbook("SampleWorkbook.xlsx")        # Variable 'WorkBook' will open the workbook for us to read. 
WorkSheet = WorkBook.sheet_by_index(0)                      # Variable 'WorkSheet' will help us navigate to 'Sheet1' of workbook.
# CellValues = WorkSheet.cell_value(0, 0)                   # Just to check syntax & help for cell_value
try:
    stringValueCount = 0                                    # This is to take care of the column header 
    sumOfAllValues = 0                                      # This variable will help us sum all values

    for i in range (0, WorkSheet.nrows):                    # This will run through all the 
        if (isinstance(WorkSheet.cell_value(i, 0), str)):   # The condition will check for any string values
            stringValueCount += 1                           # This will count number of strings if any in the entire data
        else:
            sumOfAllValues += WorkSheet.cell_value(i, 0)    # This will sum all the values in 'Column A' of 'Sheet1'
    averageVal = sumOfAllValues / (WorkSheet.nrows - stringValueCount)
    print ("Sample Mean:", averageVal)
except IndexError:                                           # This will be thrown if we interchange row and values.
    print ("list index out of range")
```
