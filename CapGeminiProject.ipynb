```python
# Code to extract employee data from an Excel sheet - First attempt - Second Iteration
# NODATA is indicated in Colour: Red

import xlrd                                        # Imports package XLRD
import time
startTime = time.time()
wb = xlrd.open_workbook("01-EmployeeData.xlsx")    # Opens the file
ws = wb.sheet_by_index(0)                          # Stores sheet with index '0'
for i in range (0, ws.nrows):
    for j in range (0, ws.ncols):
        if ws.cell_value(i, j) == '' and ws.cell_value(i, j) == '--':
            print ("\033[1;31;48mNODATA")
        else:
            print ("\033[0;30;48m", ws.cell_value(i, j))
    print ("\n")
endTime = time.time()
print (int(endTime - startTime), "Seconds")
```

```python
# Code to extract employee data from an Excel sheet - First attempt - Second Iteration
# NODATA is indicated in Colour: Red
# Adding Values to a Nested List

import xlrd                                        # Imports package XLRD
import time
startTime = time.time()
wb = xlrd.open_workbook("01-EmployeeData.xlsx")    # Opens the file
ws = wb.sheet_by_index(0)                          # Stores sheet with index '0'
myList1 = []
for i in range (0, ws.nrows):
    for j in range (0, ws.ncols):
        if ws.cell_value(i, j) == '' or ws.cell_value(i, j) == '--':
            # print ("\033[1;31;48mNODATA")
            myList1.append("NODATA")
        else:
            # print ("\033[0;30;48m", ws.cell_value(i, j))
            myList1.append(ws.cell_value(i, j))
    # print ("\n")
    # myList2.append(myList1)
# print (myList1)
# print (len(myList1))
myList2 = []
myList3 = []
x = 0
for i in range (0, len(myList1)):
    myList2 = []
    for j in range (x, ws.ncols):
        myList2.append(myList1[j])
        x += 1
    myList3.append(myList2)
# print (myList2)
print (myList3)

endTime = time.time()
print (int(endTime - startTime), "Seconds")
```
