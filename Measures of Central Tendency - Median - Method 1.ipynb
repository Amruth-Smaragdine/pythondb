```python
# This code snippet will help us find 'Median' - A measure of central tendency derived from data that's randomly generated.

import random
myList = []
for i in range (1,4):              # Odd and Even Number of values can be generated by changing the max value of the range. 
    # print (i)                     # First value in range is inclusive & last value of range is exclusive
    n = random.randint(33, 66)       # This will generate random values in range 0 to 99 and assignes them to variable 'n'
    myList.append(n)
    # print (n)                     # Intermediate print statement to check values
myList.sort()
    # print (myList)                # Intermediate print statement to check values
if len(myList) == 0 or len(myList) == 1:
    print ("There are no sufficient elements in the list")
elif len(myList)%2 == 0:
    # print (myList[int(len(myList)/2) - 1], myList[int(len(myList)/2)])        # Intermediate print statement to check values
    print ("The median of values in list: {} is:".format(myList), ((myList[int(len(myList)/2) - 1]) + myList[int(len(myList)/2)]) / 2)
else:
    print ("The median of values in list: {} is: '{}'".format(myList, myList[int(len(myList)/2)]))
```
